```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import seaborn as sns
import os
import tensorflow as tf
from tensorflow import keras as ks

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
```

```python
pokemon_df = pd.read_csv('../input/pokemon-images-and-types/pokemon.csv')
pokemon_df.head()
```

```python
pokemon_df = pokemon_df.sort_values(by=['Name'], ascending=True).reset_index(drop=True)
pokemon_df.head()
```

```python
type1 = pokemon_df['Type1'].tolist()
type2 = pokemon_df['Type2'].tolist()

types = type1 + type2
```

```python
plt.figure(figsize=(14, 6))

sns.countplot(x=types)
plt.title('Pokemons types')
plt.show()
```

```python
x1 = pokemon_df.Type1.value_counts().keys()
y1 = pokemon_df.Type1.value_counts().values
```

```python
x2 = pokemon_df.Type2.value_counts().keys()
y2 = pokemon_df.Type2.value_counts().values
```

```python
xy1 = pd.DataFrame({'x1':x1, 'y1':y1})
xy2 = pd.DataFrame({'x2':x2, 'y2': y2})
xy1.sort_values(by=['x1'],ascending=True, inplace=True)
xy2.sort_values(by=['x2'],ascending=True, inplace=True)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(30, 7))
gspec = gridspec.GridSpec(2, 2)
tp = plt.subplot(gspec[0, 1])
bt = plt.subplot(gspec[1, 1:])

tp.bar(xy1.x1,xy1.y1)
bt.bar(xy2.x2, xy2.y2)
bt.invert_yaxis()
bt.set_xticks([])
plt.tight_layout()
tp.set_title('Pokemon based on Type1 and Type2')
plt.show()
```

```python
path='../input/pokemon-images-and-types/images/images/'
img_name = sorted(os.listdir('../input/pokemon-images-and-types/images/images/'))
img_paths = []
for i in img_name:
    img_paths.append(path + i)
```

```python
pokemon_df['filepath'] = img_paths
```

```python
pokemon_df['Type2']  = pokemon_df['Type2'].fillna('None')
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
clean_types = list(set(types))
le.fit(clean_types)
le.classes_
```

```python
pokemon_df['Type1'] = le.fit_transform(pokemon_df['Type1'].values)
pokemon_df['Type2'] = le.fit_transform(pokemon_df['Type2'].values)
pokemon_df.head()
```

```python
pokemon_df = pokemon_df.drop(['Name'], axis=1)
pokemon_df.reset_index(drop=True, inplace=True)
pokemon_df.head()
```
